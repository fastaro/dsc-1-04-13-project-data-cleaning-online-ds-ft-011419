## Joining, Grouping, and Aggregating

In the cell below, subset male and female heroes into different dataframes.  Create a scatterplot of the height and weight of each hero, with weight as the y-axis.  Plot both the male and female heroes subset into each dataframe, and make the color for each point in the scatterplot correspond to the gender of the superhero.

```python
male_df=super_df.loc[super_df['Gender']=='Male']
x = male_df['Weight'].map(lambda i:abs(i))
x=x.tolist()
y = male_df['Height'].map(lambda i:abs(i))
y=y.tolist()
plt.scatter(x,y)
plt.xlabel('Weight')
plt.ylabel('Height')
plt.title('Height vs Weight')
plt.show()
```

```python
female_df=super_df.loc[super_df['Gender']=='Female']
x1 = female_df['Weight'].map(lambda i:abs(i))
x1=x1.tolist()
y1 = female_df['Height'].map(lambda i:abs(i))
y1=y1.tolist()
plt.scatter(x1,y1)
plt.xlabel('Weight')
plt.ylabel('Height')
plt.title('Height vs Weight')
plt.show()
```

```python
fig = plt.figure(figsize=(8,6))
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0,0,1,1])
male_df=super_df.loc[super_df['Gender']=='Male']
x = male_df['Weight'].map(lambda i:abs(i))
x=x.tolist()
y = male_df['Height'].map(lambda i:abs(i))
y=y.tolist()
#plt.scatter(x,y)
#plt.xlabel('Weight')
#plt.ylabel('Height')
#plt.title('Height vs Weight')



female_df=super_df.loc[super_df['Gender']=='Female']
x1 = female_df['Weight'].map(lambda i:abs(i))
x1=x1.tolist()
y1 = female_df['Height'].map(lambda i:abs(i))
y1=y1.tolist()
plt.scatter(x1,y1)
#plt.xlabel('Weight')
#plt.ylabel('Height')
#plt.title('Height vs Weight')


# Larger Figure Axes 1
ax1.scatter(x, y,color='green')
ax1.set_xlabel('Weight')
ax1.set_ylabel('Height')
#ax1.set_title('male')

# Insert Figure Axes 2
ax2.scatter(x1, y1,color='red')
ax2.set_xlabel('Weight')
ax2.set_ylabel('Height')
ax2.set_title('Male & Female Height vs Weight')

plt.show()
```
